# Import Needed Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

from sklearn.metrics import silhouette_score
```

## EDA

```python
df = pd.read_csv('/kaggle/input/wine-dataset-for-clustering/wine-clustering.csv')
df.head()
```

```python
x_axis = df.loc[:, ['Malic_Acid']]
y_axis = df.loc[:, ['Proline']]

plt.scatter(x_axis, y_axis)
```

## K-Means Model

```python
kmeans = KMeans(n_clusters=2, init='random')
kmeans.fit(df)
```

```python
pred = kmeans.predict(df)
pred
```

```python
x_axis = df.loc[:, ['Malic_Acid']]
y_axis = df.loc[:, ['Proline']]

plt.scatter(x_axis, y_axis, c=pred)
```

```python
silhouette_score(df, labels=kmeans.labels_, metric='euclidean')
```
