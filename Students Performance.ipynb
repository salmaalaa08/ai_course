```python
import pandas as pd
```

```python
# Load the data & check what's inside
df = pd.read_csv('StudentsPerformance.csv')
```

```python
# Frequency of each item in the gender column
df['gender'].value_counts()
```

```python
# Frequency of each item in the race/ethnicity column
df['race/ethnicity'].value_counts()
```

```python
# Frequency of each item in the parental level of education column
df['parental level of education'].value_counts()
```

```python
# Frequency of each item in the lunch column
df['lunch'].value_counts()
```

```python
# Frequency of each item in the test preparation course column
df['test preparation course'].value_counts()
```

```python
# A summary for math score, reading score, and writing score columns
df.describe()
```

Number of males/females in each group in race/ethnicity column

```python
df.groupby('race/ethnicity')['gender'].value_counts()
```

Number of parental level of education in each group in race/ethnicity column

```python
df.groupby('race/ethnicity')['parental level of education'].value_counts()
```

Number of students in each type of lunch in each group

```python
df.groupby('lunch')['race/ethnicity'].value_counts()
```

```python
df.groupby('lunch')['gender'].value_counts()
```

Number of students who completed and didn't complete the test preparation course in each group

```python
df.groupby('test preparation course')['race/ethnicity'].value_counts()
```

```python
df.groupby('test preparation course')['parental level of education'].value_counts()
```
